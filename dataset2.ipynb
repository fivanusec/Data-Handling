```python
import psycopg2
import matplotlib.pyplot as plt 
import json

# database handling
conn = psycopg2.connect("dbname=datasets_fun user=postgres password=postgres")
```

```python
cur = conn.cursor()
cur.execute("select d.najmoprimac, count(*) as c from dataset d group by d.najmoprimac having count(d.najmoprimac) > 1 order by c desc")
data = cur.fetchall()
data_json = dict(data)
fig, ax = plt.subplots(figsize=(20,20))
ax.bar(data_json.keys(),data_json.values(),width=0.2, edgecolor="white", linewidth=10)
plt.show()
```
